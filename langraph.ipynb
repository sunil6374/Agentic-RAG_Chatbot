```python
from typing import TypedDict
from langgraph.graph import StateGraph, START, END
```

```python
## State
```

```python
class State(TypedDict):
    name:str
    greeting:str
    age:int
```

```python
## Node 1
```

```python
def greet(state:State):
    state["greeting"]=f"Hello {state['name']}!"
    return state
```

```python
##Node 2
```

```python
def add_age(state:State):
    state["age"]=30
    return state
```

```python
## create graph
```

```python

builder=StateGraph(State)
builder.add_node("greet",greet)
builder.add_node("age",add_age)

```

```python
## Connect Them
```

```python
builder.add_edge(START,"greet")
builder.add_edge("greet","age")
builder.add_edge("age",END)
```

```python
## Compile
```

```python
graph=builder.compile()
```

```python
## Invoke
```

```python
result=graph.invoke(
    {
        'name':'Sunil'
    }
)
```

```python
print(result)
```

```python
## State
```

```python
from typing import TypedDict
from langgraph.graph import StateGraph, START, END
```

```python
class State(TypedDict):
    name:str
    age:int
    message:str
```

```python
## this is just created as a passing Node(returns state as it is) inorder to call route because only node can call
# a Route func, START cannot, it can only initialize node 
```

```python
def check_age(state:State):
    return state
```

```python
## Nodes
```

```python
def adult(state:State):
    state['message']='You are an adult'
    return state
```

```python
def minor(state:State):
    state['message']='You are a minor'
    return state
```

```python
##  route is just a routing function, not a Node
```

```python
def route(state:State):
    if state['age'] >= 18:
        return 'adult'
    return 'minor'
```

```python
## Graph
```

```python
builder=StateGraph(State)
builder.add_node("check",check_age)
builder.add_node("adult",adult)
builder.add_node("minor",minor)
```

```python
## Connect edges
```

```python
builder.add_edge(START,"check")
```

```python
## Connect Conditional Edge
```

```python

builder.add_conditional_edges("check",route)
```

```python
## Connect both nodes to END
```

```python
builder.add_edge("adult",END)
builder.add_edge("minor",END)
```

```python
graph=builder.compile()
```

```python
result=graph.invoke(
    {'name':'Sunil',
     'age': 15
     }
)
```

```python
print(result)
```
